```python
# Importing the required libraries

import pandas as pd
import numpy as np
```

```python
# importing and visualizing the data

employee = pd.read_csv('Employee.csv')
employee.head(10)
```

```python
employee.shape
```

```python
# calculating the mean salary of each department using a Groupby:

employee.groupby(['Department']).agg({'Salary':'mean'})

```

```python
# calculating the mean salary of each department using a pivot-table :

employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = ['mean'])
```

```python
# Calculating the mean salary by department and title

employee.pivot_table(index= ['Department','Title'], values = ['Salary'], aggfunc = [sum])
```

```python
# displaying the number of years of work experience and the salary for each Title


employee.pivot_table(index= ['Title'], values = ['Salary','Years'])
```

```python
# mean salary and number of employees in each department and for each job title.

employee.pivot_table(index= ['Department','Title'], values = ['Salary','Name'], aggfunc = {'Salary': 'mean','Name':len})
```

```python
# for each department and for each job title: Salary's median, minimum, maximum and standard deviation

salary_info = employee.pivot_table(index= ['Department','Title'], values = 'Salary', aggfunc = [np.mean,min,max,np.std])

salary_info


# Comments: Python doesn´t calculate de std Deviation whrn we have only 1 observation
```

```python
# The stake holders want to know for each department the number of employees and how much money is spend on salaries

employee.pivot_table(index= ['Department'], values = ['Salary','Name'], aggfunc = {'Salary': sum,'Name':len})

```

```python
# For each Department and Title, we want to know the the total years of work experience and the mean salary.


employee.pivot_table(index= ['Department','Title'], values = ['Salary','Years'], aggfunc = {'Salary':'mean','Years':sum})
```
